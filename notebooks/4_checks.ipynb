```python
df_num = df[
    [
        "type_of_failure",
        "pets",
        "carpet_score",
        "had_failure",
        "time_to_failure",
        "total_usage_time",
    ]
]
key_cols = ["type_of_failure", "pets", "carpet_score", "had_failure"]
for col in df_num.columns:
    if col == "type_of_failure":
        continue
    df_num[col] = pd.to_numeric(df_num[col])
```

```python
grouped_df = pd.DataFrame(
    columns=["type_of_failure", "pets", "carpet_score", "had_failure"]
)

agg_df_list = []

for col in df_num.columns:
    if col in key_cols:
        continue
    df_groupby = df_num.groupby(key_cols, group_keys=True)
    for agg in ["mean", "count", "max", "min", "std"]:
        colname = f"{col}_{agg}"
        print(colname)
        if agg == "mean":
            df_grouped = df_groupby.mean().reset_index()
        elif agg == "count":
            df_grouped = df_groupby.count().reset_index()
        elif agg == "max":
            df_grouped = df_groupby.max().reset_index()
        elif agg == "min":
            df_grouped = df_groupby.min().reset_index()
        elif agg == "std":
            df_grouped = df_groupby.std().reset_index()
        if df_grouped.empty:
            continue
        df_grouped.rename(columns={col: colname}, inplace=True)
        cols = key_cols + [colname]
        df_grouped = df_grouped[cols]
        # agg_df_list.append(df_grouped)
        grouped_df = grouped_df.merge(df_grouped, on=key_cols, how="outer")


# agg_df = pd.concat(agg_df_list)
# grouped_df = grouped_df.merge(agg_df, on=key_cols, how="outer")
```

```python
grouped_df.to_csv(
    "../output/aggregated_information_for_each_failure_grouped.csv", index=False
)
```
